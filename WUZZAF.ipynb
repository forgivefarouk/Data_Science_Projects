```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df =pd.read_csv('Wuzzuf_Jobs.csv')
```

```python
df.head(12)
```

```python
df.isnull().sum()
```

## what is the most Title in WUZZAF ??

```python
cnt_title = df.Title.value_counts().to_frame().reset_index()
cnt_title.columns=['Title','count']
```

```python
sns.set_style(style='darkgrid')
fig ,ax =plt.subplots(figsize=(18,7),dpi=200)
sns.barplot(data=cnt_title[:25] ,x='Title',y='count',palette='Set2')


for patch in ax.patches:
    bl = patch.get_xy()
    x = 0.5 * patch.get_width() + bl[0]
    y = 1.02 * patch.get_height() + bl[1] 
    ax.text(x,y,"%d" %(patch.get_height()),
            ha='center' , color = '#4a4a4a',
           fontsize = 12, fontfamily='serif')
    
    
for x in ['top','left','right']:
    ax.spines[x].set_visible(False)
    
    
plt.ylabel('Number of posted jobs' , fontsize = 14, fontfamily='serif');
plt.xlabel('Job Titel' , fontsize = 14, fontfamily='serif');


plt.xticks(fontsize = 12 , rotation = 90 , fontfamily='serif');

plt.title('Most Job Titles Posted on Website ' ,fontsize = 16, fontfamily='serif' );
```

## EDA for title column

```python
df['Title']=df.Title.apply(lambda x :'Sales' if 'Sales' in x or 'Teleales' in x else x)
df['Title']=df.Title.apply(lambda x :'Accountant' if 'Accountant' in x else x)
df['Title']=df.Title.apply(lambda x :'HR' if 'HR' in x else x)
df['Title']=df.Title.apply(lambda x :'Marketing' if 'Marketing' in x else x)
df['Title']=df.Title.apply(lambda x :'Data Analyst' if 'Analyst' in x or 'Analysis' in x else x)
df['Title'] = df.Title.apply(lambda x: 'Graphic designer' if 'Graphic' in x else x)
df['Title'] = df.Title.apply(lambda x: 'Bussiness Developer' if 'Bussiness Developer' in x else x)

```

```python
fig,ax=plt.subplots(figsize=(18,16),dpi=200)
df.Title.value_counts().sort_values(ascending=False).reset_index().head(25).plot(kind='bar',x='index' , ax=ax , color='gray',alpha=.7)
ax.grid(axis='y',alpha=0.6)

for patch in ax.patches:
    bl=patch.get_xy()
    x=0.5*patch.get_width()+bl[0]
    y=1.02*patch.get_height()+bl[1]
    ax.text(x,y,"%d" %(patch.get_height()),
            ha='center' , color = '#4a4a4a',
           fontsize = 12, fontfamily='serif')
    
    
for x in ['top','left','right']:
    ax.spines[x].set_visible(False)
    
    
plt.ylabel('Number of posted jobs' , fontsize = 14, fontfamily='serif');
plt.xlabel('Job Titel' , fontsize = 14, fontfamily='serif');


plt.xticks(fontsize = 12 , rotation = 90 , fontfamily='serif');

plt.title('Most Job Titles Posted on Website ' ,fontsize = 16, fontfamily='serif' );

    
```

## Skills For Data Analysis

```python
datanalysis=df[df['Title']=='Data Analyst']
datanalysis
```

## EDA for Skills column for data Analysis

```python
datanalysis['Skills']=datanalysis.Skills.apply(lambda x :'Computer Science' if 'Computer' in x or 'Software' in x or 'Information' in x else x)
datanalysis['Skills']=datanalysis.Skills.apply(lambda x :'Excel' if 'Excel' in x or 'Microsoft' in x or 'Office' in x else x)
datanalysis['Skills']=datanalysis.Skills.apply(lambda x :'Financial Analysis' if 'Financial' in x or 'Accounting' in x or 'Finance' in x else x)
datanalysis['Skills']=datanalysis.Skills.apply(lambda x :'Database' if 'Database' in x or 'SQL' in x or 'Oracle' in x else x)
#datanalysis['Skills']=datanalysis.Skills.apply(lambda x :'Statistics' if 'Statistics' in x or 'Science' in x  else x)

```

```python
fig,ax=plt.subplots(figsize=(18,6))
datanalysis.Skills.str.split(',',expand=True).stack().reset_index(level=1,drop=True).str.split().value_counts().head(40).plot(kind='bar',color='grey')

for patch in ax.patches:
    bl=patch.get_xy()
    x=0.5*patch.get_width()+bl[0]
    y=1.02*patch.get_height()+bl[1]
    ax.text(x,y,"%d" %(patch.get_height()),
            ha='center' , color = '#4a4a4a',
           fontsize = 12, fontfamily='serif')
    
    
for x in ['top','left','right']:
    ax.spines[x].set_visible(False)
```
